```python
import pandas as pd
```

```python
data = {
    'Region': ['North', 'North', 'North', 'South', 'South', 'East', 'East', 'West', 'West', 'West'],
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Furniture', 'Electronics', 'Furniture', 'Electronics', 'Furniture', 'Electronics', 'Furniture'],
    'Sales': [1000, 2000, 1500, 2500, 3000, 2000, 2500, 3000, 4000, 2000],
    'Quantity': [5, 10, 7, 12, 15, 10, 15, 20, 25, 10]
}

df = pd.DataFrame(data)
df
```

```python
pivot_sales_by_region = df.pivot_table(index='Region', values='Sales', aggfunc='sum')
pivot_sales_by_region
```

```python
pivot_avg_quantity_by_category = df.pivot_table(index='Category', values='Quantity', aggfunc='mean')
pivot_avg_quantity_by_category
```

```python
pivot_sales_quantity_by_region_category = df.pivot_table(index=['Region', 'Category'], values=['Sales', 'Quantity'], aggfunc={'Sales': 'sum', 'Quantity': 'mean'})
pivot_sales_quantity_by_region_category
```
