```python
import numpy as np
import seaborn as sns
```

```python
bimodal_data = np.concatenate((np.random.normal(loc=-5, scale=1, size=500), np.random.normal(loc=5, scale=1, size=500)))
sns.kdeplot(bimodal_data)
sns.displot(bimodal_data)
```
